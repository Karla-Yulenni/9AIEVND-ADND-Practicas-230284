*12.3 Funciónes que mandan a llamar a otras funciones y que regresan el resultado de un procedimiento*, en este caso utilizaremos una función predefinida e Python denominada *replace* que nos permite sustituir letras en una cadena de texto.

```python
def sustituir_vocales(texto):
    texto = texto.replace("A", "4")
    texto = texto.replace("a", "4")

    texto = texto.replace("E", "3")
    texto = texto.replace("e", "3")

    texto = texto.replace("I", "1")
    texto = texto.replace("i", "1")

    texto = texto.replace("O", "0")
    texto = texto.replace("o", "0")

    texto = texto.replace("U", "∪")
    texto = texto.replace("u", "∪")

    return texto
```

*Llamado*

```python
print(sustituir_vocales(saludar_nombre("Karla")))
```

**18. En este paso se calcularon los percentiles 5 y 95 de la columna MILES con el objetivo de identificar valores extremos.**
Después, se filtraron los datos para conservar únicamente los registros cuyos valores de millas se encuentran dentro de ese rango. 
Esto permite eliminar posibles outliers o datos atípicos que podrían afectar el análisis y las gráficas, dejando un conjunto 
de datos más limpio y representativo.

```python
a=df['MILES'].quantile(0.95)
b=df['MILES'].quantile(0.05)
df3=df[df['MILES']>b]
df3=df[df['MILES']<a]
df3
```

**19. Cálculo de la desviación Estándar**

para entender que tan variables o dispersas son las distancias de los viajes.

```python
df3['STD']=df.MILES.std()
df3
```

**20.Estimar el valor de Z code o Puntuación Z**

Con la finalidad de estimar la disperción de los datos, es decir que tan variados o diferentes son, utilizamos esta fórmula estadística para poder continuar con el análisis EDA (Exploratory Data Analisys)

```python
df3['Z']=(df.MILES-df.MILES.mean())/df.MILES.std()
df3
```

**23. Gráfica de clasificación de viajes por categoria**

```python
fig, ax = mat.subplots(figsize=(8,5))

grafica = sea.countplot(
    data=df,
    x='CATEGORY',
    hue='CATEGORY', 
    palette='Set2',
    legend=False,
    ax=ax
)

#Titulo
ax.set_title('Viaje por Categoría')

#Agregar etiquetas con totales 
for container in ax.containers:
    ax.bar_label(container)

mat.show()
```

**24. grafica de datos por purpose**

```python
fig, ax = mat.subplots(figsize=(10,6))

grafica = sea.countplot(
    data=df,
    y='PURPOSE',
    hue='PURPOSE',
    dodge=False,
    palette='viridis',
    legend=False
)

#Titulo
ax.set_title('Purpose of Visit (Others = Not Mentioned)')

#Agregar etiquetas con totales 
for container in ax.containers:
    ax.bar_label(container)

mat.show()
```

**25. Grafica de lugar de abordaje (START)**

```python
fig, ax = mat.subplots(figsize=(15,30))

orden = sorted(df['START'].dropna().unique())

grafica = sea.countplot(
    data=df,
    y='START',
    order=orden,
    hue='START',
    dodge=False,
    palette='viridis',
    legend=False,
    ax=ax
)

#Titulo
ax.set_title('Start Location')

#Agregar etiquetas con totales 
for container in ax.containers:
    ax.bar_label(container)

mat.show()
```

**26. Gráfica de lugar de llegada (STOP)**
